```python
#bar plot--it work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 80, 70])
color = np.array(["red","yellow", "orange", "green"])
plt.bar(fruit,quantity,color=color)
plt.xlabel("fruit")
plt.ylabel("quantity")
plt.title("Fruit vs quantity")
plt.show()
```

```python
#pie plot
percentage_of_fruit = np.array([20, 25, 30, 25])
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
myexplode = np.array([0.2, 0, 0, 0])
plt.pie(percentage_of_fruit, labels=fruit, explode = myexplode)
plt.show()
plt.legend(title="Fruit")
plt.show()
```

```python
age_of_car = np.array([4,5,6,6,7,8,9,10])
speed_of_car = np.array([150,140,130,120,110,100,90,80])
plt.scatter(age_of_car, speed_of_car)
plt.xlabel("age of car")
plt.ylabel("speed of car")
plt.title("age vs speed of car")
plt.show()
```
